# **ISOLATION FOREST ALGORITHM**

```python
import pandas as pd
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
columns_to_keep = ['V2', 'V3']
data = df[columns_to_keep]
print(data)
```

```python
print(data.isna().sum())
```

```python
data = data.dropna()
data = data.dropna(axis=1)
```

```python
print(data.isna().sum())
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.iloc[:,0], data.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.02)
clf.fit(data)
predictions = clf.predict (data)
```

```python
predictions
```

```python
import numpy as np
abc =np.where(predictions < 0)
abc
```

```python
x=data.values
```

```python
abc = np.where(predictions < 0)
plt.scatter(data.iloc[:,0], data.iloc[:,1])
plt.scatter(x[abc,0],x[abc,1],edgecolors="r")
```

## **DBSCAN ALGORITHM USING MAKE CIRCLES**

```python
from sklearn.cluster import DBSCAN
#from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.10)
dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

# **DBSCAN ALGORITHM USING DATASET**

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('creditcard.csv')
print(df.head())
```

```python
columns_to_keep = ['V23','V26','V7']
dfdata=df[columns_to_keep]
print(dfdata)
```

```python
dat=dfdata.dropna()
dat=dfdata.dropna(axis=1)
print(dat)
```

```python
# Select the first 1000 rows
sliced_df = dat.iloc[:1000]

# Display the reduced dataset
print(sliced_df)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(sliced_df)
```

```python
plt.scatter(sliced_df.iloc[:,0], sliced_df.iloc[:,1])
```

```python
dbscan.labels_
```

```python
X=sliced_df.values
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```
